# 01 - EDA: House Price Prediction

Goal of this notebook: understand the data before any cleaning or modeling.

Questions we want to answer:
- How many rows/columns do we have? What are the data types?
- Where are the missing values, and what percentage?
- How is the target `SalePrice` distributed?
- Which features are most correlated with `SalePrice`?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
pd.set_option('display.max_columns', 100)
```

## 1. Load the data

```python
df = pd.read_csv('../data/raw/train.csv')
df.shape
```

```python
df.head()
```

## 2. Overview of columns and data types

```python
df.info()
```

```python
df.describe()
```

## 3. Missing values

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_table = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
missing_table = missing_table[missing_table['missing_count'] > 0].sort_values('missing_pct', ascending=False)
missing_table
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=missing_table['missing_pct'], y=missing_table.index)
plt.title('Missing value percentage per column')
plt.xlabel('% missing')
plt.tight_layout()
plt.show()
```

## 4. Target distribution: SalePrice

```python
df['SalePrice'].describe()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['SalePrice'], kde=True, ax=axes[0])
axes[0].set_title('SalePrice distribution')

sns.histplot(np.log1p(df['SalePrice']), kde=True, ax=axes[1])
axes[1].set_title('log(SalePrice + 1) distribution')
plt.tight_layout()
plt.show()
```

## 5. Relationship between numeric features and SalePrice

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()['SalePrice'].sort_values(ascending=False)
corr.head(15)
```

```python
top_features = corr.index[1:11]
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df[top_features.tolist() + ['SalePrice']].corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Heatmap of top 10 features correlated with SalePrice')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.scatterplot(x=df['GrLivArea'], y=df['SalePrice'], ax=axes[0,0])
axes[0,0].set_title('GrLivArea vs SalePrice')

sns.boxplot(x=df['OverallQual'], y=df['SalePrice'], ax=axes[0,1])
axes[0,1].set_title('OverallQual vs SalePrice')

sns.scatterplot(x=df['TotalBsmtSF'], y=df['SalePrice'], ax=axes[1,0])
axes[1,0].set_title('TotalBsmtSF vs SalePrice')

sns.scatterplot(x=df['YearBuilt'], y=df['SalePrice'], ax=axes[1,1])
axes[1,1].set_title('YearBuilt vs SalePrice')
plt.tight_layout()
plt.show()
```
